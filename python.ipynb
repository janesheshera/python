1. Напишите скрипт, который проверяет наличие дубликатов в наборе данных и выводит информацию о количестве их обнаружения.

```python
def check_duplicates(data):
    
    # Проверка наличия дубликатов
    duplicates = data[data.duplicated()]

    # Подсчет количества дубликатов
    num_duplicates = len(duplicates)

    # Вывод информации о дубликатах
    if num_duplicates > 0:
        print(f"В наборе найдено {num_duplicates} дубликатов")
    else:
        print("Дубликаты не обнаружены.")
```

```python
check_duplicates(set1)
check_duplicates(set2)
```

2. Разработайте скрипт, который анализирует данные на наличие пропущенных и нулевых значений (там, где они не уместны) и предоставляет отчет о количестве пропусков и нулевых неуместных ячеек в каждом столбце.

```python
def check_nan_0(data):
    
    # Проверка пропусков в данных
    missing_values = data.isnull().sum()
    
    # Нулевые неуместные значения
    invalid_zeros = (data == 0).sum()
    
    # Создаем отчет о пропусках и нулевых значениях
    report = pd.DataFrame({'Пропущенные значения': missing_values, 'Нулевые неуместные значения': invalid_zeros})
    
    return report
```

```python
check_nan_0(set1)
```

```python
check_nan_0(set2)
```

4. Создайте скрипт, который идентифицирует выбросы в данных, используя статистические методы, и предоставляет отчет о найденных аномалиях. Опишите, почему вами был выбран именно этот конкретный метод/методы, какие вы видите в нем достоинства и недостатки.

```python
def identify_outliers(data):
        
    # Вычисляем межквартильный размах
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
        
    # Вычисляем верхнюю и нижнюю границы выбросов
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
        
     # Идентифицируем выбросы
    outliers = np.where((data < lower_bound) | (data > upper_bound))[0]
    return outliers
```
